```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
m = 1/10

x_unif = np.random.uniform(0, 10, 1000)
x_norm = np.random.normal(5, 0.25, 1000)
x_norm_bimodal = np.concatenate((np.random.normal(2.5, 0.1, 500), np.random.normal(7.5, 0.1, 500)))
y_unif = m * x_unif + np.random.normal(0, 0.4, 1000)
y_norm = m * x_norm + np.random.normal(0, 0.4, 1000)
y_norm_bimodal = m * x_norm_bimodal + np.random.normal(0, 0.4, 1000)
```

```python
# plot all data side by side
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(x_unif, y_unif)
plt.legend(["r = {:.2f}".format(pearsonr(x_unif, y_unif)[0])])
plt.title("Uniform")
plt.subplot(1, 3, 2)
plt.scatter(x_norm, y_norm)
plt.legend(["r = {:.2f}".format(pearsonr(x_norm, y_norm)[0])])
plt.title("Normal")
plt.subplot(1, 3, 3)
plt.scatter(x_norm_bimodal, y_norm_bimodal)
plt.legend(["r = {:.2f}".format(pearsonr(x_norm_bimodal, y_norm_bimodal)[0])])
plt.title("Bimodal Normal")
```
