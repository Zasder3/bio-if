```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from highlight_text import HighlightText
```

```python
def colorFader(c1,c2,mix=0): #fade (linear interpolate) from color c1 (at mix=0) to c2 (mix=1)
    c1=np.array(mpl.colors.to_rgb(c1))
    c2=np.array(mpl.colors.to_rgb(c2))
    return mpl.colors.to_hex((1-mix)*c1 + mix*c2)
```

```python

c1='teal' #blue
c2='red' #green
n=500

fig, ax = plt.subplots(figsize=(8, 5))
for x in range(n+1):
    ax.axvline(x, color=colorFader(c1,c2,x/n), linewidth=4) 
plt.show()
```

```python
highlight_textprops =\
[{"bbox": {"edgecolor": "orange", "facecolor": "yellow", "linewidth": 1.5, "pad": 1}},
 {"color": "#969696"}]
 
fig, ax = plt.subplots(figsize=(4, 4))  

HighlightText(x=0.5, y=0.5,
              fontsize=16,
              ha='center', va='center',
              s='The weather is <sunny>\nYesterday it was <cloudy>',
              highlight_textprops=highlight_textprops,
              ax=ax)
```

```python
cmap1 = mpl.colors.LinearSegmentedColormap.from_list(
        'unevently divided', [(0, c1), (.5, '#F2F1F1'), (1, c2)])
```

```python
aa_sequence = 'MQIFVKTLTGKTITLEVEPSDTIENVKAKIQDKEGIPPDQQRLIFAGKQLEDGRTLSDYNIQKESTLHLVLRLRGG'
data = np.random.rand(len(aa_sequence))
highlight_textprops = [{"fontfamily": "monospace", "bbox": {"pad": 0, "linewidth": 0, "facecolor": cmap1(d)}} for d in data]
# highlight_textprops = [{} for d in data]

fig, ax = plt.subplots(figsize=(10, 2))
HighlightText(x=0.01, y=0.5, 
                s="<" + "><".join(aa_sequence) + ">", 
                highlight_textprops=highlight_textprops,
                fontsize=12,
                ax=ax)
plt.xticks([]),plt.yticks([])
ax.set_facecolor("#F2F1F1")
```
