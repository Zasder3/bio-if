```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
df = pd.read_csv('../src/bio_if/data/plmbias/common_proteins_progen_esm_loglikelihood.csv')
```

```python
df.columns
```

```python
def log_formatter(x, pos):
    return f'{int(x)}'
# make plot where on left is histogram and right is CDF
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Length', kde=True, log_scale=2)
plt.gca().xaxis.set_major_formatter(FuncFormatter(log_formatter))
plt.subplot(1, 2, 2)
sns.ecdfplot(data=df, x='Length', log_scale=2)
plt.gca().xaxis.set_major_formatter(FuncFormatter(log_formatter))
```
