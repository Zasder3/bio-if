```python
import matplotlib.pyplot as plt
import torch
```

```python
esm_35m = torch.load('esm2_t12_35M_UR50D_times.pt')
esm_150m = torch.load('esm2_t30_150M_UR50D_times.pt')
esm_650m = torch.load('esm2_t33_650M_UR50D_times.pt')
```

```python
x = [32, 64, 128, 256, 512, 1024]
# with square markers make a plot
plt.plot(x, esm_35m, label='35M', marker='s')
plt.plot(x, esm_150m, label='150M', marker='s')
plt.plot(x, esm_650m, label='650M', marker='s')
plt.xlabel('Training Dataset Size')
plt.ylabel('Time (s)')
plt.yscale('log')
plt.xscale('log', base=2)
plt.title('ESM2 IF Computation vs. Training Dataset Size (256 Queries)')
plt.legend()
# add more numbers to x-axis and y-axis
plt.xticks(x, [str(i) for i in x])
plt.yticks([15*60, 30*60, 60*60, 2*60*60], ['15m', '30m', '1h', '2h'])
```
